```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(data.iloc[:,1:11])
StandardScaler(copy=True,with_mean=True,with_std=True)
scaled_data=scaler.transform(data.iloc[:,1:11])
scaled_data.shape
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(scaled_data)
x_pca=pca.transform(scaled_data)
scaled_data.shape
x_pca.shape
```

```python
from __future__ import annotations
print(data['country'].unique())
from sklearn.preprocessing import LabelEncoder
label_en=LabelEncoder()
a=['country']
for i in np.arange(len(a)):
    data[a[i]]=label_en.fit_transform(data[a[i]])

plt.figure(figsize=(8,6))
#color = ['red' if i==0  else 'green' for i in data['country']]
annotations=data['country']
x=x_pca[:,0]
y=x_pca[:,1]
plt.scatter(x_pca[:,0],x_pca[:,1],c=data['country'],cmap='Greens')
for i, label in enumerate(annotations):
    plt.annotate(label, (x[i], y[i]))

plt.xlabel('HEALTH')
plt.ylabel('ECONOMY')
plt.show()
plt.legend()
rslt_df = data[data['country'] == 22]
print(rslt_df)
rslt_df = data[data['country'] == 35]
print(rslt_df)
```
